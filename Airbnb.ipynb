```python
#Adding NY map underneath
```

```python
import urllib
from PIL import Image

urllib.request.urlretrieve('https://upload.wikimedia.org/wikipedia/commons/e/ec/Neighbourhoods_New_York_City_Map.PNG',
                           'NY_Map.png')
i = Image.open('NY_Map.png')
plt.imshow(i, zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])
ax = plt.gca()
affordable_df.plot(
    ax=ax,
    zorder=1,
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```
